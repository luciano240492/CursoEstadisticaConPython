Realizacin de grficos simples consumiendo el set de datos de temperaturas por ao de la pgina del GBA
Objetivo
- Realizar un grfico de la evolucin de las temperaturas mximas y mnimas por ao tomando la media aritmtica de los 12 meses de cada ao

```python
import pandas as pd
temp = pd.read_csv("historico_temperaturas.csv")
```

```python
temp.columns
```

```python
media = temp['mxima'].mean()
mediana = temp['mxima'].median()
media,mediana
```

```python
# Recortar el 10%
from scipy.stats import trim_mean
trim_mean(temp['mxima'],0.1)
```

```python
aos1 = []
for i in range(1991, 2021):
    aos1.append(i)
temp_por_ao1 = []
temp_por_ao2 = []
for i in range(1991,2021):
    media_max = temp.loc[temp['ao']== i]['mxima'].mean()
    temp_por_ao1.append(media_max)
for i in range(1991,2021):
    media_min = temp.loc[temp['ao']== i]['mnima'].mean()
    temp_por_ao2.append(media_min)


data1 = {'ao': aos1, 'max':temp_por_ao1,'min':temp_por_ao2 }
data1
```

```python
import pandas as pd
df = pd.DataFrame(data=data1)
df
```

```python
import matplotlib.pyplot as plt

fig4, ax4 = plt.subplots(figsize= (12,6))
ax4.plot(df['ao'],df['max'], label = 'Media max', color = "red", marker= "*")
ax4.plot(df['ao'],df['min'], label = 'Media min', color = "blue", marker= "_")
plt.legend(loc='best')
plt.grid()
```
